```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
temperature = pd.read_csv("tas_timeseries_monthly_cru_1901-2021_KEN.csv", skiprows=2)
rain = pd.read_csv("pr_timeseries_monthly_cru_1901-2021_KEN.csv", skiprows=2)
temperature = temperature.rename(columns={"Unnamed: 0": "Years"})
rain = rain.rename(columns={"Unnamed: 0": "Years"})
```

```python
display(temperature.head())
display(rain.head())
```

```python

temp_melt = pd.melt(temperature, id_vars=['Years'], var_name='Month', value_name='Temperature - (Celsius)')

# Create a new 'date' column by combining 'year' and 'month' columns
temp_melt['date'] = pd.to_datetime(temp_melt['Years'].astype(str) + '-' + temp_melt['Month'], format='%Y-%b')

# Set the 'date' column as the index
temp_melt.set_index('date', inplace=True)

# Drop the 'id', 'year', and 'month' columns
temp_melt.drop(['Years', 'Month'], axis=1, inplace=True)

# Resample the data at a monthly frequency and forward-fill missing values
temp_melt = temp_melt.resample('M').ffill()

# Reset the index to get 'date' column back as a regular column
temp_melt.reset_index(inplace=True)

# Extract 'year' and 'month' from the 'date' column
temp_melt['Year'] = temp_melt['date'].dt.year
temp_melt['month'] = temp_melt['date'].dt.strftime('%b')

# Reorder the columns to match the desired format
temp_melt = temp_melt[['Year', 'month', 'Temperature - (Celsius)']]
```

```python

rain_melt = pd.melt(rain, id_vars=['Years'], var_name='Month', value_name='Rainfall - (MM)')

# Create a new 'date' column by combining 'year' and 'month' columns
rain_melt['date'] = pd.to_datetime(rain_melt['Years'].astype(str) + '-' + rain_melt['Month'], format='%Y-%b')

# Set the 'date' column as the index
rain_melt.set_index('date', inplace=True)

# Drop the 'id', 'year', and 'month' columns
rain_melt.drop(['Years', 'Month'], axis=1, inplace=True)

# Resample the data at a monthly frequency and forward-fill missing values
rain_melt = rain_melt.resample('M').ffill()

# Reset the index to get 'date' column back as a regular column
rain_melt.reset_index(inplace=True)

# Extract 'year' and 'month' from the 'date' column
rain_melt['Year'] = rain_melt['date'].dt.year
rain_melt['month'] = rain_melt['date'].dt.strftime('%b')

# Reorder the columns to match the desired format
rain_melt = rain_melt[['Year', 'month', 'Rainfall - (MM)']]
```

```python
rain_melt
```

```python
final_df = temp_melt.merge(rain_melt)
final_df = final_df.rename(columns={"month": "Month Average"})
final_df = final_df[final_df['Year'] >= 1990]
final_df
```

```python


final_df.set_index("Year", inplace=True)
```

```python
final_df.to_csv("Kenya-climate-1990-2021.csv")
```
